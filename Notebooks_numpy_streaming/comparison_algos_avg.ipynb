```python
from config_simulation import (
    algo,
    simulation_linear_regression,
    simulation_linear_regression_toeplitz,
    simulation_linear_regression_toeplitz_diag,
    simulation_logistic_regression,
    simulation_logistic_regression_toeplitz,
    simulation_logistic_regression_toeplitz_diag,
    simulation_geometric_median,
    simulation_p_means,
    eval_covtype,
    N,
    n,
)

from functools import partial
batch_size = 32
averaged=True
multiply_lr = 'default'
optimizer_list = [
    # SGD
    partial(algo.SGD, averaged=averaged),

    # AdaGrad
    partial(algo.AdaGrad, averaged=averaged),
    
    # SNA 
    partial(algo.SNA, averaged=averaged),
    
    # USNA 
    partial(algo.USNA, averaged=averaged),

    # Streaming 
    partial(algo.SGD, averaged=averaged, batch_size=batch_size, multiply_lr=multiply_lr),
    partial(algo.AdaGrad, averaged=averaged, batch_size=batch_size, multiply_lr=multiply_lr),
    partial(algo.SNA, averaged=averaged, batch_size=batch_size, multiply_lr=multiply_lr),
    partial(algo.USNA, averaged=averaged, proj=True, batch_size=batch_size, multiply_lr=multiply_lr),
]
```
