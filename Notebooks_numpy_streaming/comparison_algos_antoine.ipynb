```python
from config_simulation import (
    algo,
    simulation_linear_regression,
    simulation_linear_regression_toeplitz,
    simulation_linear_regression_toeplitz_diag,
    simulation_logistic_regression,
    simulation_logistic_regression_toeplitz,
    simulation_logistic_regression_toeplitz_diag,
    simulation_geometric_median,
    simulation_p_means,
    eval_covtype,
)

from functools import partial
batch_size_power = 1.0
multiply_lr = 'default'
optimizer_list = [
    # WASGD
    partial(algo.SGD, averaged=True),

    # WAAdaGrad
    partial(algo.AdaGrad, averaged=True, true_ada=True),
    
    # WASNA
    partial(algo.SNA, averaged=True),
    
    # UWASNA
    partial(algo.USNA, averaged=True),

    # Streaming WASNA and UWASNA
    partial(algo.SGD, averaged=True, batch_size_power=batch_size_power, multiply_lr=multiply_lr),
    partial(algo.AdaGrad, averaged=True, true_ada=True, batch_size_power=batch_size_power, multiply_lr=multiply_lr),
    partial(algo.SNA, averaged=True, batch_size_power=batch_size_power, multiply_lr=multiply_lr),
    partial(algo.USNA, averaged=True, proj=True, batch_size_power=batch_size_power, multiply_lr=multiply_lr),
]
```
