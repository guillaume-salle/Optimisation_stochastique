```python
def step(x, h, lr, lim):
    if h * lr < lim:
        return x * (1 - lr*h)**2 + 2 * lr
    else:
        return x

def loss(a, h):
    return abs(a - 1/h)
```

```python
a_0 = 1
h = 10
lims = [1, 1.5, 1.9, 1.95]
n = 40

import matplotlib.pyplot as plt

# store the evolution of the loss for each gamma

for lim in lims:
    losses = []
    a = a_0
    n_iter = 1
    for i in range(n):
        current_loss = loss(a, h)
        losses.append(current_loss)
        lr = n_iter ** -0.75
        a = step(a, h, lr, lim)
        n_iter += 1
    plt.plot(losses, label=f'lim={lim}')
plt.legend()
        
        
        
```
